```python
import cv2 
import matplotlib.pyplot as plt
```

```python
img=cv2.imread('img.jpg')
plt.imshow(img)
```

```python
#cropping
imgcrop=img[1000:,250:1250]
plt.imshow(imgcrop)
```

```python
img_gray=cv2.cvtColor(imgcrop,cv2.COLOR_BGR2GRAY)
plt.imshow(img_gray,cmap='gray')
```

```python
img_gray_small=cv2.resize(img_gray,(25,25))
plt.imshow(img_gray_small,cmap='gray')
```

```python
print(img_gray_small)
```

```python
imcolor=cv2.cvtColor(imgcrop,cv2.COLOR_BGR2RGB)
plt.imshow(imcolor)
```

```python
crop=imcolor[-3:,-3:]
print(crop)
```
